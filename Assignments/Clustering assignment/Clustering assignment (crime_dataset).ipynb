Hierarchical Clustering

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"E:\data science excelr\(7)clustering\crime_data.csv")
df.head()
```

```python
df= df.rename({'Unnamed: 0':'City'},axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
x = df.drop(columns=["City"])
```

```python
x = sc.fit_transform(x)
x
```

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x,method='single')
dendrogram = hierarchy.dendrogram(lk)
```

```python
lk = hierarchy.linkage(x,method='complete') 
dendrogram = hierarchy.dendrogram(lk)
```

```python
lk = hierarchy.linkage(x,method='average') 
dendrogram = hierarchy.dendrogram(lk)
```

```python
lk = hierarchy.linkage(x,method='weighted') 
dendrogram = hierarchy.dendrogram(lk)
```

```python
lk = hierarchy.linkage(x,method='centroid') 
dendrogram = hierarchy.dendrogram(lk)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=4)
y=hc.fit_predict(x)
```

```python
df['Cluster']=y
```

```python
df
```

```python
df.groupby('Cluster').agg('mean')
```

K means


```python
from sklearn.cluster import KMeans
```

```python
km=KMeans()
y=km.fit_predict(x)
```

```python
km.cluster_centers_
```

```python
x=pd.DataFrame(x)
x['Cluster']=y
x
```

```python
wcss=[]

for i in range(1,25):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,25),wcss)
```
