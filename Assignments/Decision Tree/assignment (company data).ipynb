# Problem Statement

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv(r"E:\data science excelr\(14)Decision tree\Company_Data.csv")
df
```

```python
#Data Description

df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Sales'].mean()
```

```python
df['High']=df.Sales.map(lambda x: 1 if x > 7.49
                        else 0)
df
```

```python
le= LabelEncoder()
```

```python
df['ShelveLoc'] = le.fit_transform(df['ShelveLoc'])
df['Urban'] = le.fit_transform(df['Urban'])
df['US'] = le.fit_transform(df['US'])
```

```python
df
```

```python
df.info
```

```python
x= df.iloc[:,1:11]
y=df['High']
```

```python
x
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x,y,train_size=0.75, random_state=143)
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

```python
plt.rcParams['figure.dpi'] = 250

tree.plot_tree(dt, feature_names=x.columns, 
               class_names=['Sale is High','Sale is Low'],
              filled=True)
plt.show()
```

```python
dt = DecisionTreeClassifier(criterion='gini')
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

```python
plt.rcParams['figure.dpi'] = 250

tree.plot_tree(dt, feature_names=x.columns, 
               class_names=['Sale is High','Sale is Low'],
              filled=True)

plt.show()
```
