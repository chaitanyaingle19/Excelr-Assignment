```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv(r"E:\data science excelr\(9)Association Rules\book.csv")
df
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
score = apriori(df, min_support=0.1,use_colnames=True)
score
```

```python
model = association_rules(score,metric='lift')
```

```python
model
```

```python
plt.scatter(model['support'],model['confidence'])
plt.show()
```
