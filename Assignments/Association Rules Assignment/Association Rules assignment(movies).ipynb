```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df = pd.read_csv(r"E:\data science excelr\(9)Association Rules\my_movies.csv")
df
```

```python
df1= df.iloc[:,5:]
df1
```

```python
score = apriori(df1,min_support=0.1,use_colnames=True)
score
```

```python
model = association_rules(score,metric='lift')
model
```

```python
plt.scatter(model['support'],model['confidence'])
plt.show()
```

```python
model[model['lift']>1]
```
