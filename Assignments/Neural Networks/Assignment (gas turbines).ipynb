```python
import pandas as pd 
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report,r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"E:\data science excelr\(16)Neural Networks\gas_turbines.csv")
df
```

```python
x = df.iloc[:,df.columns != 'TEY']
y=df['TEY']
```

```python
sc  = StandardScaler()
x= sc.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1,)
```

```python
ann=Sequential()
ann.add(Dense(units=30,activation='relu'))
ann.add(Dense(units=20,activation='relu'))
ann.add(Dense(units=1))
ann.compile(optimizer='adam',loss='mse')
ann.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=250)
```

```python
loss=pd.DataFrame(ann.history.history)
loss.plot()
```

```python
ypred= ann.predict(xtest)
```

```python
print(r2_score(ytest,ypred))
```

```python
early= EarlyStopping(monitor='val_loss',mode='min',patience=20)
```

```python
ann=Sequential()
ann.add(Dense(units=30,activation='relu'))
ann.add(Dense(units=20,activation='relu'))
ann.add(Dense(units=1))
ann.compile(optimizer='adam',loss='mse')
ann.fit(xtrain,ytrain,validation_data=(xtest,ytest),callbacks=[early],epochs=250)
```

```python
early_loss= pd.DataFrame(ann.history.history)
early_loss.plot()
```

```python
ypred= ann.predict(xtest)
```

```python
print(r2_score(ytest,ypred))
```

```python
ann=Sequential()
ann.add(Dense(units=30,activation='relu'))
ann.add(Dropout(rate=0.5))
ann.add(Dense(units=20,activation='relu'))
ann.add(Dropout(rate=0.5))
ann.add(Dense(units=1))
ann.compile(optimizer='adam',loss='mse')
ann.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=250)
```

```python
drop_loss= pd.DataFrame(ann.history.history)
drop_loss.plot()
```
